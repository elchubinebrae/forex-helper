```python
import yfinance as yf
import numpy as np
from datetime import date,datetime, timedelta
import pandas as pd
from datetime import datetime, timedelta
import time


list_of_currency_pairs = [
    "EURUSD=X",  # Euro to US Dollar
    "USDJPY=X",  # US Dollar to Japanese Yen
    "GBPUSD=X",  # British Pound Sterling to US Dollar
    "AUDUSD=X",  # Australian Dollar to US Dollar
    "USDCAD=X",  # US Dollar to Canadian Dollar
    "USDCHF=X",  # US Dollar to Swiss Franc
    "NZDUSD=X",  # New Zealand Dollar to US Dollar
    "EURGBP=X",  # Euro to British Pound Sterling
    "EURJPY=X",  # Euro to Japanese Yen
    "GBPJPY=X",  # British Pound Sterling to Japanese Yen
    "USDZAR=X",  # US Dollar to Whatever that is
    "CADCHF=X",  # Canadian Dollar to Swiss Franc
    "CADJPY=X",  # Canadian Dollar to Japanese Yen
    "AUDJPY=X",  # Australian Dollar to Japanese Yen
    "GBPAUD=X",  # Pound to Aussie
    "EURAUD=X",
    "GBPCAD=X",
    "AUDCHF=X",
    "CHFJPY=X",
    "AUDCAD=X",
    "USDNOK=X",
    "EURCHF=X",
    "NZDJPY=X", "USDGBP=X", "NZDCAD=X", "USDTRY=X", "EURTRY=X"
]


def todays_date():
  # Get today's date
  today = datetime.now().date() + timedelta(days = 1)
  return today

def _30_minute_old_date():
  # Calculate a date 60 days ago
  date_60_days_ago = todays_date() - timedelta(days=59)
  return date_60_days_ago

def one_minute_old_date():
  # Calculate a date 60 days ago
  date_60_days_ago = todays_date() - timedelta(days=6)
  return date_60_days_ago

def six_months_ago():
  date_6_months_ago = todays_date() - timedelta(days=6*30)
  return date_6_months_ago

def stock_trender_monthly(tickle):
  # Fetch and preprocess data
    data = yf.download(tickle, start= six_months_ago(), end= todays_date(), interval="1mo")
    return data

def stock_trender_weekly(tickle):
    data = yf.download(tickle, start= six_months_ago(), end= todays_date(), interval="1wk")
    return data

def stock_trender_daily(tickle):
    data = yf.download(tickle, start= six_months_ago(), end= todays_date(), interval="1d")
    return data

def stock_trender_1_hourly(tickle):
    data = yf.download(tickle, start= six_months_ago(), end= todays_date(), interval="1h")
    return data

def stock_trender_30_min(tickle):
    data = yf.download(tickle, start= _30_minute_old_date(), end= todays_date(), interval="30m")
    return data

def stock_trender_15_min(tickle):
    data = yf.download(tickle, start= _30_minute_old_date(), end= todays_date(), interval="15m")
    return data

def stock_trender_10_min(tickle):
    data = yf.download(tickle, start= _30_minute_old_date(), end= todays_date(), interval="10m")
    return data

def stock_trender_5_min(tickle):
    data = yf.download(tickle, start= _30_minute_old_date(), end= todays_date(), interval="5m")
    return data

def stock_trender_1_min(tickle):
    data = yf.download(tickle, start= one_minute_old_date(), end= todays_date(), interval="1m")
    return data

def slope_values(your_series):
  x_values = np.arange(len(your_series))

  # Use numpy's polyfit to perform linear regression and get the slope
  slope, intercept = np.polyfit(x_values, your_series, 1)

  # If the slope is positive, it's trending upwards; if negative, it's trending downwards
  if slope > 0:
      end_of_statement = f'The values are trending upwards. Slope is {slope}'
  elif slope < 0:
      end_of_statement = f"The values are trending downwards. Slope is {slope}"
  else:
      end_of_statement = f"The values are not trending either upwards or downwards (flat line). {slope}"

  return end_of_statement

def list_slope_values(your_series):
  x_values = np.arange(len(your_series))

  # Use numpy's polyfit to perform linear regression and get the slope
  slope, intercept = np.polyfit(x_values, your_series, 1)
  return slope

def calculate_ema_and_ribbon(data, closing_price_column='Close'):
    # Calculate 200EMA
    data['200EMA'] = data[closing_price_column].ewm(span=200, min_periods=0).mean()

    # Define EMAs for the ribbon
    ribbon_periods = [14, 20, 26, 32, 38, 44, 50, 56]

    # Calculate EMAs for the ribbon
    for period in ribbon_periods:
        column_name = f'{period}EMA'
        data[column_name] = data[closing_price_column].ewm(span=period, min_periods=0).mean()

    return data
```

```python
monthly_slope = []
weekly_slope = []
daily_slope = []
hourly_slope = []
thirty_min = []
fifteen_min = []
five_min = []
one_min = []

for pair in list_of_currency_pairs:
  ticker_name = pair
  monthly_slope.append(list_slope_values(stock_trender_monthly(ticker_name).Close))
  weekly_slope.append(list_slope_values(stock_trender_weekly(ticker_name).Close))
  daily_slope.append(list_slope_values(stock_trender_daily(ticker_name).Close))
  hourly_slope.append(list_slope_values(stock_trender_1_hourly(ticker_name).Close))
  thirty_min.append(list_slope_values(stock_trender_30_min(ticker_name).Close))
  fifteen_min.append(list_slope_values(stock_trender_15_min(ticker_name).Close))
  five_min.append(list_slope_values(stock_trender_5_min(ticker_name).Close))
  one_min.append(list_slope_values(stock_trender_1_min(ticker_name).Close))
  time.sleep(3)

data = {
    'Name': list_of_currency_pairs,
    'Monthly': monthly_slope,
    'Weekly': weekly_slope,
    'Daily': daily_slope,
    'Hourly': hourly_slope,
    'Thirty Min': thirty_min,
    'Fifteen Min': fifteen_min,
    'Five Min': five_min,
   # 'One Min': one_min
}

# Convert the dictionary to a pandas DataFrame
df = pd.DataFrame(data)#
columns_to_check = df.columns[df.columns != 'Name']

# Create a new column 'All_Positive' that checks if all selected columns are positive
df['All_Positive'] = (df[columns_to_check] > 0).all(axis=1)
df['All_Negative'] = (df[columns_to_check] < 0).all(axis=1)


# Clear the output
clear_output(wait=False)
print("Code finished")
```
